```python


data_path = '/Users/andre/Documents/UCL/DoctorSignature/data/processed/training_metrics'
data = []
for f in os.listdir(data_path):
    if '.csv' in f:
        df = pd.read_csv(os.path.join(data_path, f))
        df=df[:5]
        df.loc[:, 'filename'] = f
        data.append(df)
```

```python
df_metrics = pd.concat(data)
```

```python
df_metrics.loc[:, 'metric'] = df_metrics.filename.str.split('-').str[-1].str[:-4]
df_metrics.loc[:, 'model'] = df_metrics.filename.str.split('-').str[1].str[:3]
df_metrics.loc[:, 'type'] = df_metrics.filename.str.split('-').str[-3].str.split('_').str[-1].str.capitalize()
df_metrics.loc[:, 'new'] = df_metrics.filename.str.contains('2dd477d0-4a7a-4821-b877-ec735dc6536f')


df_gru = df_metrics[df_metrics.model == 'GRU']
df_sig = df_metrics[(df_metrics.model == 'SIG') & (~df_metrics.new)]
df_sig_n = df_metrics[(df_metrics.model == 'SIG') & (df_metrics.new)]
df_sig_n.loc[:,'model'] = 'SIG_n'
```

```python
df_gru.loc[:, 'relative'] = df_gru.loc[:, 'Wall time'] - df_gru.loc[:, 'Wall time'].min()
df_sig.loc[:, 'relative'] = df_sig.loc[:, 'Wall time'] - df_sig.loc[:, 'Wall time'].min()
#df_sig_n.loc[:, 'relative'] = df_sig_n.loc[:, 'Wall time'] - df_sig_n.loc[:, 'Wall time'].min()

df_metrics_pro = pd.concat([df_gru, df_sig]) #df_sig_n
```

```python
%matplotlib notebook
df_metrics_pro.metric.uniquefig, ax = plt.subplots(figsize=(10,10))
df_plot = df_metrics_pro[df_metrics_pro.metric == 'loss']
for ((m, typ), df_p) in df_plot.groupby(['model', 'type']):
    df_p.plot(x='relative', y='Value', label=m+' '+typ, ax=ax, 
              marker='o' if m == 'SIG' else '^', alpha=0.7,linewidth=3, markersize=10, 
              linestyle='-' if typ == 'Validation' else ':')
    ax.set_xlabel('Relative wall time (seconds)')
    ax.set_ylabel('Loss')
```

```python
%matplotlib notebook
df_metrics_pro.metric.uniquefig, ax = plt.subplots(figsize=(10,10))
df_plot = df_metrics_pro[df_metrics_pro.metric == 'auc']
for ((m, typ), df_p) in df_plot.groupby(['model', 'type']):
    df_p.plot(x='relative', y='Value', label=m+' '+typ, ax=ax, 
              marker='o' if m == 'SIG' else '^', alpha=0.7,linewidth=3, markersize=10, 
              linestyle='-' if typ == 'Validation' else ':')
    ax.set_xlabel('Relative wall time (seconds)')
    ax.set_ylabel('AUROC')
    
    
    
    
    
```
